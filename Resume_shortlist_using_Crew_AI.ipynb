```python
from crewai import Agent, Task, Crew
from langchain_community.chat_models import ChatLiteLLM
import os
os.environ["GOOGLE_API_KEY"] = "AIzaSyBFejKRK1aTlv9RiYTyX4wy0PyHtnONwGA"
llm = ChatLiteLLM(
    model="gemini/gemini-1.5-flash",
    temperature=0,
    api_key = os.environ["GOOGLE_API_KEY"]
)
```

```python
resume_text = """
Name: Sritharan S
Skills: Python, Tensorflow, Machine Learning, pandas, Deep Learning
Experience: 1 years at BIM Vision Designs
Education: B.Tech Artificial INtelligence and Data Science
"""
```

```python
evaluator = Agent(
    role = "data analyst",
    goal = "evaluate resume",
    backstory = "Expert in talent assessment and AI hiring standards.",
    verbose = True,
    llm = llm
)
```

```python
writer = Agent(
    role = "Reply writer",
    goal = "Write clear replies to candidates",
    backstory = "An HR pro who crafts polite offers/reject messaages.",
    verbose = True,
    llm = llm

)
```

```python
#Task 1: Evaluation
evaluation_task = Task(
    description=f"""check if the following resume is fir for an AI/ML Engineer role.
Respond with YES or NO and a short reason:\n\n{resume_text}""",
    agent = evaluator,
    expected_output = "YES or NO with short reason, one reply"
)
```

```python
#Task 2: write reply
reply_task = Task(
    description="""Using the evaluation result, write:
- An acceptance message if YES
- A polite rejection if NO""",
    agent = writer,
    context = [evaluation_task],
    expected_output = "A professional message to send to the candidate, one reply"
)
```

```python
crew = Crew(
    agents = [evaluator, writer],
    tasks = [evaluation_task, reply_task],
    verbose = True
)
result = crew.kickoff()
print("\n Final Message to Candidate:\n")
print(result)
```
